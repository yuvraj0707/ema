```python
##Erosion Dilation Opening Closing
import numpy as np
import matplotlib.pylab as plt
import cv2
from google.colab.patches import cv2_imshow
from google.colab import files

img=cv2.imread("trial.jpg")
img=cv2.resize(img,(300,300))
cv2_imshow(img)
img=cv2.cvtColor(img,cv2.COLOR_BGR2GRAY)
m,n=img.shape

#Structuring Elemtent
P=5
SE=np.ones((P,P),dtype=np.uint8)
c=(P-1)//2

#Erosion
imgEr=np.zeros((m,n),dtype=np.uint8)
for i in range(c,m-c):
  for j in range(c,n-c):
    temp=img[i-c:i+c+1,j-c:j+c+1]
    product=SE*temp
    imgEr[i,j]=np.min(product)
plt.imshow(imgEr,cmap="gray")

#Dilation
imgDi=np.zeros((m,n),dtype=np.uint8)
for i in range (c,m-c):
  for j in range(c,n-c):
    temp=img[i-c:i+c+1,j-c:j+c+1]
    product=SE*temp
    imgDi[i,j]=np.max(product)
plt.imshow(imgDi,cmap="gray")

#Opening -Erosion +Dilation

imgOp=np.zeros((m,n),dtype=np.uint8)
for i in range (c,m-c):
  for j in range(c,n-c):
    temp=imgEr[i-c:i+c+1,j-c:j+c+1]
    product=SE*temp
    imgOp[i,j]=np.max(product)


plt.imshow(imgOp,cmap="gray")

#Closing - Dilation + Erosion
imgCl=np.zeros((m,n),dtype=np.uint8)
for i in range(c,m-c):
  for j in range(c,n-c):
    temp=imgDi[i-c:i+c+1,j-c:j+c+1]
    product=SE*temp
    imgCl[i,j]=np.min(product)

plt.imshow(imgCl,cmap="gray")
```

```python
##Experiment 1(b)
import numpy as np
import cv2
import google.colab
from google.colab.patches import cv2_imshow
img1=cv2.imread("file1.jpg")
img1=cv2.resize(img1,(300,300))
cv2_imshow(img1)
print("\n")
img2=cv2.imread("file2.jpg");
img2=cv2.resize(img2,(300,300));
cv2_imshow(img2)
#Addition
img_add=cv2.add(img1,img2);
cv2_imshow(img_add);
img_add2=img1+img2;
cv2_imshow(img_add2)
img_add3=cv2.addWeighted(img1,0.8,img2,0.2,1)
cv2_imshow(img_add3);
#Subtract
img_sub=cv2.subtract(img1,img2)
cv2_imshow(img_sub)
img_sub1=img1-img2
cv2_imshow(img_sub1)
#Multiplication
img_mul=cv2.multiply(img1,img2)
cv2_imshow(img_mul)
img_mul1=img1*img2
cv2_imshow(img_mul1)
#Division
img_div=cv2.divide(img1,img2)
cv2_imshow(img_div)
#Transpose
img_trans=cv2.transpose(img1)
cv2_imshow(img_trans)
#Logical Operations
img_and=cv2.bitwise_and(img1,img2)
cv2_imshow(img_and)
img_or=cv2.bitwise_or(img1,img2)
cv2_imshow(img_or)
img_xor=cv2.bitwise_xor(img1,img2)
cv2_imshow(img_xor)
img_not=cv2.bitwise_not(img1)
cv2_imshow(img_not)
```
